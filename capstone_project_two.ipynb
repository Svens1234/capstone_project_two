```python
#Create a scatterplot exploring the relationship between RT Critic reviews and RT User reviews.
```

```python
plt.figure(figsize=(10,4),dpi=150)
sns.scatterplot(data=all_sites,x='RottenTomatoes',y='RottenTomatoes_User')
plt.xlim(0,100)
plt.ylim(0,100)
```

```python
#Create a new column based off the difference between critics 
#ratings and users ratings for Rotten Tomatoes. Calculate this 
#with RottenTomatoes-RottenTomatoes_User
```

```python
all_sites['Rotten_Diff']  = all_sites['RottenTomatoes'] - all_sites['RottenTomatoes_User']
```

```python
all_sites['Rotten_Diff']
```

```python
#Calculate the Mean Absolute Difference between RT scores and RT User scores as described above.
```

```python
all_sites['Rotten_Diff'].apply(abs).mean()
```

```python
#Plot the distribution of the differences between RT Critics Score 
#and RT User Score. There should be negative values in this distribution 
#plot. Feel free to use KDE or Histograms to display this distribution.
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.histplot(data=all_sites,x='Rotten_Diff',kde=True,bins=25)
plt.title("RT Critics Score minus RT User Score");
```

```python
#What are the top 5 movies users rated higher than critics on average
```

```python
all_sites.nsmallest(5,'Rotten_Diff')
```

```python
all_sites.nsmallest(5,'Rotten_Diff')['FILM']
```

```python
all_sites.nsmallest(5,'Rotten_Diff')[['FILM','Rotten_Diff']]
```

```python
#Now show the top 5 movies critics scores higher than users on average.
```

```python
all_sites.nlargest(5,'Rotten_Diff')[['FILM','Rotten_Diff']]
```
